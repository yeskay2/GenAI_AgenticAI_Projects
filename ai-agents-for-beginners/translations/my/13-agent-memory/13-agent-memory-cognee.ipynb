## ğŸ› ï¸ á€¡á€€á€°á€¡á€Šá€®á€•á€±á€¸á€á€±á€¬ Function: Session History á€€á€­á€¯á€€á€¼á€Šá€·á€ºá€›á€¾á€¯á€›á€”á€º

á€’á€® utility function á€€ Redis á€™á€¾á€¬á€á€­á€™á€ºá€¸á€†á€Šá€ºá€¸á€‘á€¬á€¸á€á€²á€· á€…á€€á€¬á€¸á€á€­á€¯á€„á€ºá€¸á€á€™á€­á€¯á€„á€ºá€¸á€€á€­á€¯ á€…á€…á€ºá€†á€±á€¸á€”á€­á€¯á€„á€ºá€…á€±á€•á€«á€á€šá€ºá‹ á€’á€«á€€ á€¡á€±á€¬á€€á€ºá€•á€«á€¡á€á€½á€€á€º á€¡á€á€¯á€¶á€¸á€á€„á€ºá€•á€«á€á€šá€º -  
- Session á€€á€­á€¯á€…á€®á€™á€¶á€á€”á€·á€ºá€á€½á€²á€™á€¾á€¯ Debugging á€œá€¯á€•á€ºá€›á€”á€º  
- á€…á€€á€¬á€¸á€á€­á€¯á€„á€ºá€¸á€á€½á€± Cache á€œá€¯á€•á€ºá€‘á€¬á€¸á€á€¬á€€á€­á€¯ á€¡á€á€Šá€ºá€•á€¼á€¯á€›á€”á€º  
- Agent á€¡á€á€½á€€á€º á€›á€›á€¾á€­á€”á€­á€¯á€„á€ºá€á€²á€· context á€€á€­á€¯ á€”á€¬á€¸á€œá€Šá€ºá€›á€”á€º  


````python
async def show_history(session_id: str) -> None:
    # Let's check the cache directly
    cache_engine = get_cache_engine()
    if cache_engine:
        # Try to get history directly from cache
        user_id = str(user.id) if hasattr(user, 'id') else None
        if user_id:
            history_entries = await cache_engine.get_latest_qa(user_id, session_id, last_n=10)
            print(f"\nDirect cache query for user_id={user_id}, session_id={session_id}:")
            print(f"Found {len(history_entries)} entries")
            if history_entries:
                for i, entry in enumerate(history_entries, 1):
                    print(f"\nEntry {i}:")
                    print(f"  Question: {entry.get('question', 'N/A')[:100]}...")
                    print(f"  Answer: {entry.get('answer', 'N/A')[:100]}...")
        else:
            print("No user_id available")
````
