## 📋 Προαπαιτούμενα & Ρύθμιση

### Απαιτήσεις Συστήματος

Πριν ξεκινήσετε, βεβαιωθείτε ότι έχετε:

1. **Περιβάλλον Python**
   - Python 3.9 ή νεότερη έκδοση
   - Εικονικό περιβάλλον (συνιστάται)
   
2. **Redis Cache** (Απαραίτητο για τη Διαχείριση Συνεδριών)
   - Τοπικό Redis: `docker run -d -p 6379:6379 redis`
   - Ή χρησιμοποιήστε μια διαχειριζόμενη υπηρεσία Redis
   
3. **Πρόσβαση σε LLM API**
   - Κλειδί API OpenAI ή άλλων παρόχων (δείτε [τεκμηρίωση](https://docs.cognee.ai/setup-configuration/llm-providers))

4. **Ρύθμιση Βάσης Δεδομένων**
   - Δεν απαιτείται ρύθμιση από προεπιλογή. Το Cognee χρησιμοποιεί βάσεις δεδομένων βασισμένες σε αρχεία (LanceDB και Kuzu)
   - Προαιρετικά, μπορείτε να ρυθμίσετε το Azure AI Search ως κατάστημα διανυσμάτων (δείτε [τεκμηρίωση](https://github.com/topoteretes/cognee-community/tree/main/packages/vector/azureaisearch))

### Ρύθμιση Περιβάλλοντος

Δημιουργήστε ένα αρχείο `.env` στον κατάλογο του έργου σας με τις παρακάτω μεταβλητές:

```ini
# LLM Configuration (Required)
LLM_API_KEY=your-openai-api-key-here

# Cache Configuration (Required for Sessions)
CACHING=true  # Must be enabled for session history

```

```
            │

│                                          │
│  ┌────────────────────────────────────┐  │
│  │ Knowledge Graph in Graph Database  │  │
│  └────────────────────────────────────┘  │

            │                      ▲   
            ▼                      │(optional)
┌────────────────┐           ┌────────────────┐

│                │           │    (Redis)     │

        ▲
        │
┌──────────────────────────┐
│          Agents          │

```

## 📁 Διαμόρφωση Καταλόγων Αποθήκευσης

Το Cognee χρησιμοποιεί δύο ξεχωριστούς καταλόγους για τις λειτουργίες του:
- **Data Root**: Αποθηκεύει τα εισαγόμενα έγγραφα και τα επεξεργασμένα δεδομένα
- **System Root**: Περιέχει τη βάση δεδομένων του γραφήματος γνώσης και τα μεταδεδομένα του συστήματος

Θα δημιουργήσουμε απομονωμένους καταλόγους για αυτό το σεμινάριο ως εξής:


````python
DATA_ROOT = Path('.data_storage').resolve()
SYSTEM_ROOT = Path('.cognee_system').resolve()

DATA_ROOT.mkdir(parents=True, exist_ok=True)
SYSTEM_ROOT.mkdir(parents=True, exist_ok=True)

cognee.config.data_root_directory(str(DATA_ROOT))
cognee.config.system_root_directory(str(SYSTEM_ROOT))

print(f"Data root: {DATA_ROOT}")
print(f"System root: {SYSTEM_ROOT}")
````

## 🔐 Μέρος 3: Ρύθμιση Διαχείρισης Συνεδρίας

### Ενεργοποίηση Μνήμης Συνομιλίας

Η διαχείριση συνεδρίας είναι κρίσιμη για τη διατήρηση του πλαισίου σε πολλαπλές αλληλεπιδράσεις. Εδώ θα:

1. **Αρχικοποιήσουμε το Πλαίσιο Χρήστη**: Δημιουργήσουμε ή ανακτήσουμε ένα προφίλ χρήστη για την παρακολούθηση της συνεδρίας
2. **Ρυθμίσουμε τη Μηχανή Cache**: Συνδεθούμε με το Redis για την αποθήκευση του ιστορικού συνομιλιών
3. **Ενεργοποιήσουμε Μεταβλητές Συνεδρίας**: Ρυθμίσουμε μεταβλητές πλαισίου που παραμένουν σταθερές μεταξύ ερωτημάτων

> ⚠️ **Σημαντικό**: Απαιτείται το Redis να είναι σε λειτουργία και το `CACHING=true` στο περιβάλλον σας


````python
from cognee.modules.users.methods import get_default_user
from cognee.context_global_variables import set_session_user_context_variable 
from cognee.infrastructure.databases.cache import get_cache_engine

user = await get_default_user()
await set_session_user_context_variable(user)
print(f"Using user id: {getattr(user, 'id', 'unknown')}")

cache_engine = get_cache_engine()
if cache_engine is None:
    raise RuntimeError('Cache engine is not available. Double-check your cache configuration.')
print('Session cache is ready.')
````

## 🛠️ Βοηθητική Συνάρτηση: Προβολή Ιστορικού Συνεδρίας

Αυτή η βοηθητική συνάρτηση μας επιτρέπει να επιθεωρούμε το ιστορικό συνομιλιών που αποθηκεύεται στο Redis. Είναι χρήσιμη για:
- Εντοπισμό σφαλμάτων στη διαχείριση συνεδριών
- Επαλήθευση ότι οι συνομιλίες αποθηκεύονται προσωρινά
- Κατανόηση του διαθέσιμου πλαισίου για τον πράκτορα


````python
async def show_history(session_id: str) -> None:
    # Let's check the cache directly
    cache_engine = get_cache_engine()
    if cache_engine:
        # Try to get history directly from cache
        user_id = str(user.id) if hasattr(user, 'id') else None
        if user_id:
            history_entries = await cache_engine.get_latest_qa(user_id, session_id, last_n=10)
            print(f"\nDirect cache query for user_id={user_id}, session_id={session_id}:")
            print(f"Found {len(history_entries)} entries")
            if history_entries:
                for i, entry in enumerate(history_entries, 1):
                    print(f"\nEntry {i}:")
                    print(f"  Question: {entry.get('question', 'N/A')[:100]}...")
                    print(f"  Answer: {entry.get('answer', 'N/A')[:100]}...")
        else:
            print("No user_id available")
````
